# Concatenando ambos ficheros 

```python
cat_df = pd.read_csv("categorical_data.csv",
                         keep_default_na=False, na_values=[""])

num_df = pd.read_csv("numeric_data.csv",
                         keep_default_na=False, na_values=[""])
```

```python
# Colocando los __DataFrames__ de lado a lado
union_ficheros = pd.merge(left=cat_df,right=num_df)
```

```python
print(union_ficheros.head(10))
```

# Graficas de Dispersión (Ficheros combinados) 

```python
# Evaluacion jugadores de los primeros 30 clubes en el ranking
# (Vissel kobe equipo de menor nivel del top 30 dado que tiene 4 jugadores por debajo del umbral de 60 y tan solo 1 por arriba)

plt.scatter(union_ficheros["Overall"],union_ficheros["Club"].astype(str))
plt.margins(x=0.1)
plt.xlabel('Overall')
plt.ylabel('Club')
plt.title('Dispersion overall del top 30 de Clubes')  
plt.ylim(0, 30)  
plt.show()
```

```python
# Evaluación jugadores de los primeros 10 clubes en el ranking 
# (peor jugador por equipo Man City, mejor jugador, empate entre Juventus y FC Barcelona)

plt.scatter(union_ficheros["Overall"],union_ficheros["Club"].astype(str))
plt.margins(x=0.1)
plt.xlabel('Overall')
plt.ylabel('Club')
plt.title('Dispersion relativa del top 10 de Clubes')
plt.ylim(0, 10)             
plt.show()
```

```python
# Evaluacion jugadores de los primeros 10 selecciones en el ranking
# (La seleccion Francesa es la más equilibrada, con todos sus jugadores en el rango de 50:90 Overall)

plt.scatter(union_ficheros["Overall"],union_ficheros["Nationality"].astype(str))
plt.margins(x=0.1)
plt.xlabel('Overall')
plt.ylabel('Nationality')
plt.title('Dispersion relativa del top 10 de Selecciones')
plt.ylim(0, 10)             
plt.show()
```
